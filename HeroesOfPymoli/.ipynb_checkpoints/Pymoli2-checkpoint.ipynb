```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

```python
player_count = len(purchase_data_df["SN"].value_counts())
#print(player_count)

player_count_df=pd.DataFrame({"Total Players":[player_count]})
player_count_df
```

```python

# Run basic calculations to obtain number of unique items, average price, etc.

# Create a summary data frame to hold the results

# Optional: give the displayed data cleaner formatting

# Display the summary data frame
unique_data = len(purchase_data_df["Item ID"].value_counts())
print(unique_data)

avg_price = purchase_data_df["Price"].mean()
print (avg_price)

# No.of Purchases
noofpurchase_data = len(purchase_data_df["Purchase ID"].value_counts())
print(noofpurchase_data )

# Total Revenue
totRevenue_data = purchase_data_df["Price"].sum()
print(totRevenue_data)
```

```python
purchase_analysis_df=pd.DataFrame({"Number of Unique Items":[unique_data],
                                  "Average Price":[avg_price],
                                   "Number of Purchases":[noofpurchase_data],
                                   "Total Revenue":[totRevenue_data]
                                  
                                  })

purchase_analysis_df["Average Price"]=purchase_analysis_df["Average Price"].map("{:.2f}".format)
purchase_analysis_df
```

```python
exclnull_df = purchase_data_df.dropna(how='any')

totcnt=len(exclnull_df["SN"].value_counts())

gender_counts = exclnull_df["Gender"].value_counts()
gender_counts.head()

```

```python
#Gender Demographics
exclnull_df = purchase_data_df.dropna(how='any')

grouped_exclnull_df = exclnull_df.groupby(['Gender'])

#total number of players, excluding duplicates
cnt=grouped_exclnull_df["SN"].value_counts().count()
#print(cnt)
#grouped_exclnull_df.SN.unique()
#grouped_exclnull_df['SN'].value_counts()

ngrp=exclnull_df.groupby("Gender").SN.nunique()



percent_players= (ngrp / cnt) * 100


newSummary_df=pd.DataFrame({"Total Count":ngrp,
                          "Percentage of Players":percent_players
                            })
newSummary_df
```

```python
# Purchasing Analysis (Gender)
 
# Run basic calculations to obtain purchase count, avg. purchase price, 
#avg. purchase total per person etc. by gender



# Optional: give the displayed data cleaner formatting

# Display the summary data frame
exclnull_df2 = purchase_data_df.dropna(how='any')

grouped_exclnull_df2 = exclnull_df2.groupby(['Gender'])

#purchase count
purchase_cnt= grouped_exclnull_df2['Purchase ID'].count()

#avg. purchase price
avgpurchase_price = grouped_exclnull_df2['Price'].mean()

#total purchase value
totpurchase_val = grouped_exclnull_df2['Price'].sum()

ngrp=exclnull_df2.groupby('Gender')['SN'].nunique()

#avg. purchase total per person etc.
avgpurtot_person = totpurchase_val/ngrp
#print(avgpurtot_person)

# Create a summary data frame to hold the results

SummaryGender_PurAnalysis_df = pd.DataFrame({"Purchase Count":purchase_cnt,
                                             "Average Purchase Price":avgpurchase_price,
                                             "Total Purchase Value" : totpurchase_val,
                                             "Avg total Purchase per Person":avgpurtot_person
    
    
}

)


SummaryGender_PurAnalysis_df["Average Purchase Price"]=SummaryGender_PurAnalysis_df["Average Purchase Price"].map("{:.2f}".format)

SummaryGender_PurAnalysis_df["Total Purchase Value"]=SummaryGender_PurAnalysis_df["Total Purchase Value"].map("{:.2f}".format)

SummaryGender_PurAnalysis_df["Avg total Purchase per Person"]=SummaryGender_PurAnalysis_df["Avg total Purchase per Person"].map("{:.2f}".format)

SummaryGender_PurAnalysis_df
```

```python

exclnull_df2 = purchase_data_df.dropna(how='any')

# Establish bins for ages
age_bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,129]

categ_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut()
#creating an Age_summary column in the data frame
exclnull_df2["Age_Summary"] = pd.cut(exclnull_df2["Age"],age_bins, labels=categ_names)
exclnull_df2

age_summary_df = exclnull_df2.groupby("Age_Summary")
age_summary_df.head()
# Calculate the numbers and percentages by age group

totplayers_agegrp = age_summary_df['SN'].nunique()
perc_agegrp = (totplayers_agegrp / cnt) * 100

# Create a summary data frame to hold the results

ageSummaryDf_final = pd.DataFrame({"Total Count":totplayers_agegrp ,
                                   "Percentage of Players":perc_agegrp
    
    
})

# Optional: round the percentage column to two decimal points

ageSummaryDf_final["Percentage of Players"]=ageSummaryDf_final["Percentage of Players"].map("{:.2f}%".format)

ageSummaryDf_final
# Display Age Demographics Table
```
